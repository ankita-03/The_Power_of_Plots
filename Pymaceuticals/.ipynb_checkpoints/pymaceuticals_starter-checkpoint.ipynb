## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single data set
#use pd.merge 

data_set = pd.merge(mouse_metadata, study_results, how='left', on='Mouse ID')
data_set
# Display the data table for preview
```

```python
# Checking the number of mice.
    # use unique or nunique
num_mice = data_set['Mouse ID'].nunique()
num_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
data_set[data_set.duplicated(['Mouse ID','Timepoint'])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
data_set[data_set['Mouse ID']=='g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean = data_set[data_set['Mouse ID']!='g989']
clean
```

```python
# Checking the number of mice in the clean DataFrame.
clean['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

grp = clean.groupby('Drug Regimen')['Tumor Volume (mm3)']

pd.DataFrame({
    'mean': grp.mean(),
    'median':grp.median(),
    'var':grp.var(),
    'std':grp.std(),
    'sem':grp.sem()
})
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
grp.agg(['mean','median','var','std','sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse
grp2 = clean.groupby(['Mouse ID'])['Timepoint'].max()
tp = pd.DataFrame(grp2)

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merge1 = pd.merge(clean, grp2, on =('Mouse ID','Timepoint'), how = 'right')
merge1.head(31)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
data1 = []
data2 = []
data3 = []
data4 = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for index, row in merge1.iterrows():
    if row['Drug Regimen'] == treatments[0]:
        data1.append(row['Tumor Volume (mm3)'])
    if row['Drug Regimen'] == treatments[1]:
        data2.append(row['Tumor Volume (mm3)'])
    if row['Drug Regimen'] == treatments[2]:
        data3.append(row['Tumor Volume (mm3)'])
    if row['Drug Regimen'] == treatments[3]:
        data4.append(row['Tumor Volume (mm3)'])
    
    
    # add subset 
pd.DataFrame({
    treatments[0]: data1,
    treatments[1]: data2,
    treatments[2]: data3,
    treatments[3]: data4
    
})
    
    # Determine outliers using upper and lower bounds
# stuck at this part   
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
x = pd.DataFrame({
    treatments[0]: data1,
    treatments[1]: data2,
    treatments[2]: data3,
    treatments[3]: data4
    
}) 
plt.boxplot(x)
plt.title("Final Tumor Volume of each Mouse across Four Treatment Drugs")
plt.xticks([1,2,3,4], treatments)
plt.figure(figsize=[10,10])
plt.savefig("fig 3")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# using b128 from list above
z = clean[clean['Mouse ID']=='b128']
z
xaxis = z['Timepoint']
yaxis = z['Tumor Volume (mm3)']
plt.title("Capomulin treated on b128")
plt.plot(xaxis, yaxis, markersize = 10)
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoints (days)")
plt.figure(figsize=(12,12))
plt.savefig("fig 4")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
z1 = clean[clean['Drug Regimen']=='Capomulin']
z1

avg_tumor = z1.groupby(['Mouse ID']).mean()
x2 = avg_tumor['Tumor Volume (mm3)']
y2 = avg_tumor['Weight (g)']
plt.scatter(x2,y2)
plt.title("Avg Tumor Volume vs Mouse Weight")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.savefig("fig 5")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

line = st.linregress(x2,y2)
line

slope = 0.7426941826944178
y_int = -10.148863164111322

lin_reg = slope*(x2) + y_int

plt.scatter(x2,y2)
plt.plot(x2, lin_reg)
plt.title("Liner Regression Model")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.savefig("fig 6")
plt.show()
```
